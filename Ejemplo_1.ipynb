```python
#impor dpylot module with the plt elias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pnd
import scipy as sp

#create a figure or dashboard with axes

fig, ax = plt.subplots()

#draw references points
ax.scatter(x = [1,2,3], y = [3,2,1])

#For save dashboard
plt.savefig('diagrama.png')

#For show
plt.show()


```

```python
fig,ax = plt.subplots()
ax.barh([1,2,3],[3,2,1])
plt.show()
```

```python
fig, ax  = plt.subplots()
x = np.random.normal (5, 1.5, size = 1000)
ax.hist(x, range(0, 11))
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie([2,2,2,2])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,0,0.5])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.fill_between([1,2,3,4],[1,2,0,0.5])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.violinplot([1,2,1,2,3,4,3,3,5,7])
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.random.random((16,16))
ax.imshow(x)
plt.show()
```
